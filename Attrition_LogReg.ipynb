# ATTRITION DATASET ANALYSIS 

```python
# loading necessary imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
# Reading the dataset
at = pd.read_csv('Attrition.csv')
```

```python
# Checking first 5 records
at.head()
```

```python
at.isnull().sum()  # There is no missing values in the dataset.
```

```python
# Description of the dataset.
at.describe(include = "all")
```

### Performing Label Encoding on the categorical data

```python
# Using Label encoding for conversion of object datatypes.
le = LabelEncoder()
at[at.select_dtypes(include = ['object']).columns] = at[at.select_dtypes(include = ['object']).columns].apply(le.fit_transform)
```

```python
at.head()
```

### Check for Multicollinearity

#### 1. Check for multi-collinearity using correlation matrix

```python
at.corr()
```

```python
at_x = at.drop('Attrition', axis = 1)
at_y = at.iloc[:,1]
```

#### 2. Check for multi-collinearity using Variance Inflation factor.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['vif_values'] = [variance_inflation_factor(at_x.values, i) for i in range(at_x.shape[1])]
vif['variables'] = at_x.columns
vif
```

## Applying Logistis Regression Model

Consider the Attrition column as the target variable

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
at_x_train,at_x_test, at_y_train, at_y_test = train_test_split(at_x, at_y, test_size = 0.2, random_state = 42)
```

```python
log = LogisticRegression()
log.fit(at_x_train, at_y_train)
pred_y = log.predict(at_x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

### Confusion Matrix

```python
tab1 = confusion_matrix(at_y_test, pred_y)  # Here predicted comes to the top and actual in the left side
tab1
```

### Accuracy of the model

```python
Accuracy = tab1.diagonal().sum() * 100 / tab1.sum()
Accuracy
```

### Probability scores for binary classes "Yes" : 1 and "No": 0 for the target (Attrition)

```python
# Ordering probability of attrition rate (1) in descending order.
#(As there is no ID column in dataset, index can be treated as ID).

pred_proba_df = pd.DataFrame(log.predict_proba(at_x))
pred_proba_df.sort_values(pred_proba_df.index[1], ascending = False)
```

```python
log.intercept_  # This shows the intercept of the Logistic Regression model fitted.
```

```python
# Co-efficients for the independent variables
beta_df = pd.DataFrame({'Variables':list(at_x.columns), 'beta':list(log.coef_.flatten())}) 
beta_df
```

### AUROC curve

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
roc_auc_value = roc_auc_score(at_y_test, pred_y)
roc_auc_value
```

```python
# Plotting AUROC curve

pred_value_prob = pd.DataFrame(log.predict_proba(at_x_test))
pred_value_prob = pred_value_prob.iloc[:,1] # only selecting the +ve (1/yes) column probabilities.

fpr, tpr, threshold = roc_curve(at_y_test, pred_value_prob)

plt.figure(figsize=(10,10))
plt.plot(fpr,tpr, color = 'r', label="LogisticRegression(area=%.2f)" %roc_auc_value)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('AUROC ON THE ATTRITION')
plt.legend(loc=1)
plt.grid(linestyle = '--')
```

### F1 score

```python
# Lets plot Precision-Recall curve

from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
```

```python
pred_value = log.predict(at_x_train)
pred_value_prob = log.predict_proba(at_x_train)
pred_value_prob = pd.DataFrame(pred_value_prob)
pred_value_prob1 = pred_value_prob.iloc[:,1]
precision, recall, thresholds = precision_recall_curve(at_y_train, pred_value_prob1)
f1 = f1_score(at_y_train, pred_value)
f1
```

#### From the above score for f1, it is clear that either of the two classes(1 or 0) has a class_Imbalance problem

## Here we can see that due to lesser number of positive cases in attrition, class 1 does not 

## get proper representation

```python
at.Attrition.value_counts()  # So we need to up-sample the class 1.
```

```python
# Percentage of weightage of class 0
per = 1233*100/1470
per
```

```python
# Percentage of weightage of class 1
per1 = 237*100/1470
per1
```

### Splitting Dataset into train and test before resolving the Class-Imbalance problem

```python
at_train, at_test = train_test_split(at, test_size = 0.2, random_state = 10)
```

```python
a = at_train[at.Attrition > 0]
a.head()
```

```python
a.shape
```

```python
at_train = pd.concat([at_train,a,a,a])
```

```python
at_train.Attrition.value_counts()
```

```python
at_train.Attrition.value_counts(1)*100   # Percentage weightage of each class after resolving the class-imbalance problem.
```

#### From the above observation, it is clear that the class imbalance problem is almost resolved, so refit the model on balanced data.

```python
at_x_train = at_train.drop('Attrition', axis = 1) # independent variables for training set
at_y_train = at_train.loc[:,'Attrition']  # Target variable for training set
```

```python
at_x_test = at_test.drop('Attrition', axis = 1)  # independent variables for testing set
at_y_test = at_test.loc[:,'Attrition']   # Target variable for testing set
```

### Fitting the Logistic Regression model on dataset after resolving Class-Imbalance problem

```python
log.fit(at_x_train, at_y_train)
```

```python
pred_y = log.predict(at_x_test) # predicting on the test data.
```

```python
# confusion Matrix

tab_new = confusion_matrix(at_y_test, pred_y)  # Here predicted comes to the top and actual in the left side
tab_new
```

```python
# Accuracy of the model

Accuracy = tab1.diagonal().sum() * 100 / tab1.sum()
Accuracy
```

### Classsification Report

```python
from sklearn import metrics

print('Report : ')
print(metrics.classification_report(at_y_test, pred_y))
```

### AUROC curve

```python
roc_auc_value = roc_auc_score(at_y_test, pred_y)
roc_auc_value
```

```python
# Plotting AUROC curve after improving class imbalance problem

pred_value_prob = pd.DataFrame(log.predict_proba(at_x_test))
pred_value_prob = pred_value_prob.iloc[:,1] # only selecting the +ve (1/yes) column probabilities.

fpr, tpr, threshold = roc_curve(at_y_test, pred_value_prob)

plt.figure(figsize=(10,10))
plt.plot(fpr,tpr, color = 'r', label="LogisticRegression(area=%.2f)" %roc_auc_value)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('AUROC ON THE ATTRITION')
plt.legend(loc=1)
plt.grid(linestyle = '--')
```

#### Here we can observe that the auroc acore has improved from 65% to 75% after resolving class-imbalance problem.

```python
# Precision-recall after improving class imbalance problem

pred_value = log.predict(at_x_train)
pred_value_prob = log.predict_proba(at_x_train)
pred_value_prob = pd.DataFrame(pred_value_prob)
pred_value_prob1 = pred_value_prob.iloc[:,1]
precision, recall, thresholds = precision_recall_curve(at_y_train, pred_value_prob1)
f1 = f1_score(at_y_train, pred_value)
f1
```

```python
# Plotting Precision-recall after improving class imbalance problem

plt.figure(figsize=(10,10))
plt.plot(recall,precision, color = 'r', label="LogisticRegression(area=%.2f)" %f1)
plt.xlabel('recall')
plt.ylabel('precision')
plt.title('Precision-Recall ON THE ATTRITION')
plt.legend(loc=1)
plt.grid(linestyle = '--')
```

#### Here we can observe that the f1 acore has also improved from 45% to 72% after resolving class-imbalance problem.
